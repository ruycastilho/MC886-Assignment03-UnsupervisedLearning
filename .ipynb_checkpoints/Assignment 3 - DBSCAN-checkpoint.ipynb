## Importação dos Dados

```python
import pandas as pd
import numpy as np
import os

def load_health():
    txt_path = os.path.join("health-dataset", "health.txt")
    with open(txt_path) as f:
        data = f.readlines()[1:]
    parsed = []
    for line in data:
        parsed.append(line.replace('\n', '').split('|'))
    return np.array(parsed)


def load_bags():
    csv_path = os.path.join("health-dataset", "bags.csv")
    return pd.read_csv(csv_path, header=None)

def load_word2vec():
    csv_path = os.path.join("health-dataset", "word2vec.csv")
    return pd.read_csv(csv_path)
```

## DBSCAN

```python
# Bag of Words
bag = np.array(load_bags())
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics.pairwise import cosine_similarity, cosine_distances, euclidean_distances
import numpy as np
from scipy import sparse, spatial
from math import log

def dbscanClustering(eps, min_samples, metric):
    dbscan = DBSCAN(eps=eps, min_samples=min_samples, metric=metric, n_jobs=-1)

    return dbscan.fit(bag)
```

## Performance Metrics

### Silhouette Coefficient

```python
from sklearn.metrics.pairwise import cosine_similarity, cosine_distances, euclidean_distances
from sklearn.metrics import silhouette_score

def silhouette(data, labels, metric):
    return silhouette_score(data, labels, metric=metric)
```

### Calinski Harabaz

```python
from sklearn.metrics import calinski_harabaz_score

def calinski_harabaz(data, labels):
    return calinski_harabaz_score(data, labels)
```

## Tests

```python
results = []
metric = 'cosine'

for eps in range(1,10):
    for min_samples in range(2,31):
        clustered = dbscanClustering(eps/10, min_samples, metric)
        if ( len(np.unique(clustered.labels_)) > 1 and len(np.unique(clustered.labels_)) <= len(bag) -1 ):
            silhouette_score_dbscan = silhouette(bag, clustered.labels_, metric)
            calinski_score_dbscan = calinski_harabaz(bag, clustered.labels_)

            results.append((eps, min_samples, silhouette_score_dbscan, calinski_score_dbscan))
```

```python
import numpy
numpy.save('results', results)
```

```python
import numpy
numpy.savetxt('results.txt', results)
```

```python
eps_plot = np.array([i[0] /10 for i in results])
min_samples_plot = np.array([i[1] for i in results])
silhouette_plot = np.array([i[2] for i in results])
calinski_plot = np.array([i[3] for i in results])
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.tri as mtri

fig = plt.figure(figsize=plt.figaspect(0.3))
x = eps_plot.flatten()
y = min_samples_plot.flatten()
triang = mtri.Triangulation(x, y)
ax = fig.add_subplot(1, 2, 2, projection='3d')
ax.plot_trisurf(triang, silhouette_plot, cmap='viridis')
ax.view_init(30, 235)
plt.xlabel("eps", fontsize=12)
plt.ylabel("min-samples", fontsize=12)
ax.set_zlabel('Silhouette Score', fontsize=12)
plt.savefig('silhouette3d.png', dpi=300)
plt.show();
```
