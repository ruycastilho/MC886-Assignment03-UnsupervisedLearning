## Importação dos Dados

```python
import pandas as pd
import numpy as np
import os

def load_health():
    txt_path = os.path.join("health-dataset", "health.txt")
    with open(txt_path) as f:
        data = f.readlines()[1:]
    parsed = []
    for line in data:
        parsed.append(line.replace('\n', '').split('|'))
    return parsed


def load_bags():
    csv_path = os.path.join("health-dataset", "bags.csv")
    return pd.read_csv(csv_path)

def load_word2vec():
    csv_path = os.path.join("health-dataset", "word2vec.csv")
    return pd.read_csv(csv_path)
```

## Geração de matrizes de Features e Target

### Setup inicial dos dados

```python
import numpy as np

#carrega os dados
data = load_bags()
data_train = data[0:int(len(data)*0.9)]
data_test = data[int(len(data)*0.9):]

def getTrainSet():
    return data_train

def getTestSet():
    return data_test
```

## Gerador de conjuntos de Cross Validation

```python
# Cross Validation Generation 

from sklearn.model_selection import train_test_split, KFold

# Retorna um gerador de indices
def generate_sets(TRAINING_DATA,type='kfold'):

    # Cross validation using train_test_split
    if (type == 'split'):
       return train_test_split(TRAINING_DATA,test_size=0.1,random_state=0)

    # Cross validation using K-Fold
    # K = 5, Shuffle = true, Seed = 21
    elif (type == 'kfold'):
        kfold_seed = 21

        kfold = KFold(n_splits=5, shuffle=True, random_state=kfold_seed)
        return kfold.split(TRAINING_DATA)
```

## Elbow Method

```python
from sklearn.cluster import KMeans
wcss = []

trainData = getTrainSet()

for i in range(1, 500):
    kmeans = KMeans(n_clusters = i, init = 'random',n_jobs=-1,random_state=21)
    kmeans.fit(trainData)
    print(i,kmeans.inertia_)
    wcss.append(kmeans.inertia_)  
```

```python
len(wcss)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1, 500), wcss)
plt.title('Elbow Method')
plt.xlabel('Clusters')
plt.ylabel('WSS') #within cluster sum of squares
plt.show()
```

## Metrics

### Silhouette Coefficient

```python
from sklearn.metrics.pairwise import cosine_similarity, cosine_distances, euclidean_distances
from sklearn.metrics import silhouette_score

def silhouette(data, labels, metric):
    return silhouette_score(data, labels, metric=metric)
```

### Calinski Harabaz

```python
from sklearn.metrics import calinski_harabaz_score

def calinski_harabaz(data, labels):
    return calinski_harabaz_score(data, labels)
```
